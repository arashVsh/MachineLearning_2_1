## Load libraries

```python
import torch
import torch.nn as nn
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from ArtificialNeuralNetwork import trainANN, testANN
from Configurations import configuration
import time
```

### Device

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

### Load dataset

```python
transform = transforms.Compose(
    [transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))]
)
train_data = datasets.MNIST(
    root="./data", train=True, transform=transform, download=True
)
test_data = datasets.MNIST(
    root="./data", train=False, transform=transform, download=True
)
train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
test_loader = DataLoader(test_data, batch_size=32)
```

### Create validation dataset

```python
from torch.utils.data import random_split, DataLoader

# Split the test set into training and validation
total_samples = len(train_loader.dataset)
val_size = total_samples // 8  # 1/8 of the training set
train_size = total_samples - val_size

train_subset, val_subset = random_split(train_loader.dataset, [train_size, val_size])

# Create data loaders for training and validation
train_loader = DataLoader(train_subset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_subset, batch_size=32, shuffle=False)
```

### Choose configuration, train, and evaluate

```python
for choice in range(30):
    print('Results of choice {}:'.format(choice))
    start_time = time.time() # Start time

    layers, activation = configuration(choice)
    epochs = 10
    model = trainANN(layers, activation, epochs, train_loader, val_loader, device)
    testANN(model, test_loader, device)

    end_time = time.time() # End time
    running_time = end_time - start_time
    print(f"Total running time: {running_time:.2f} seconds")

```
